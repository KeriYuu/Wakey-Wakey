```python
def num2hex(num):
    a = hex(int(int(num)/16))[-1]# 商
    b = hex(int(int(num)%16))[-1]# 余数
    return(a+b)


def main(colors):
    reslist = []    
    for color in colors:
        res = '#'
        for i in color:    
            res = res + num2hex(i)
        reslist.append(res)
    return reslist

if __name__=='__main__':
    colors = [(233, 208, 48), (111, 6, 126), (234, 94, 28), (124, 198, 224), (200, 0, 41), 
              (185, 187, 119), (119, 117, 118), (55, 64, 60), (218, 112, 167), (47, 99, 171),
              (232, 120, 86), (71, 17, 140), (233, 157, 37), (132, 9, 118), (221, 246, 70),
              (134, 0, 17), (126, 174, 48), (106, 45, 18), (230, 0, 28), (36, 51, 23)]
    reslist = main(colors)
    print(reslist)
```
